```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('/content/nasa93dem.csv')
```

```python
symCols = [col for col in data.columns if col.strip()[0].islower() and data[col].dtypes == 'O']
```

```python
data.head()
```

```python
le = LabelEncoder()
for sym in symCols:
  tempCol = le.fit_transform(data[sym])
  data[sym] = tempCol.copy()
```
